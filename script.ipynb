```python
df = pd.DataFrame(list(dic.items()),
                   columns=['Name', 'Change'])

df.to_excel("raw_data.xlsx")
```

```python
IDs = list(df['Name'])
Nodes = {}
for ID in IDs:
    Nodes[ID] = ID
    
df_Nodes = pd.DataFrame(list(Nodes.items()),
                   columns=['Id', 'Label'])

df_Nodes.to_excel("Nodes.xlsx", index=False)
            
#df_test2 = pd.DataFrame(Test_data2,columns=['Sorce', 'Target', 'Weight'])

df_positive = df[df['Change'] > 0]
df_negative = df[df['Change'] < 0]
average_market_return = float(df.mean())


i_1 = 0 
i_2 = 0
n = 0

all_negative = []
all_positive = []

while i_1 < len(df_negative) - 1:
    while i_2 < len(df_negative) - 1:
        
        if n < i_2 +1:
            change_1 = df_negative.iloc[i_1][1]
            change_2 = df_negative.iloc[i_2 + 1][1]
            average_portfolio_return = (change_1+change_2)/2
            weight = np.var([average_market_return, average_portfolio_return])/np.var([change_1, change_2])
            
            if weight > 10: 
                weight = 10
                all_negative.append((df_negative.iloc[i_1][0], df_negative.iloc[i_2 + 1][0], weight))
                i_2 +=1
            else:
                all_negative.append((df_negative.iloc[i_1][0], df_negative.iloc[i_2 + 1][0], weight))
                i_2 +=1
                
        
        else:
            i_2 +=1
               
    n += 1
    i_1 += 1
    i_2 = 0

i_1 = 0 
i_2 = 0
n = 0

while i_1 < len(df_positive) - 1:
    while i_2 < len(df_positive) - 1:
        
        if n < i_2 +1:
            change_1 = df_positive.iloc[i_1][1]
            change_2 = df_positive.iloc[i_2 + 1][1]
            average_portfolio_return = (change_1+change_2)/2 
            weight = np.var([average_market_return, average_portfolio_return])/np.var([change_1, change_2])
            
            if weight > 10:
                weight = 10
                all_positive.append((df_positive.iloc[i_1][0], df_positive.iloc[i_2 + 1][0], weight))
                i_2 +=1
            else:
                all_positive.append((df_positive.iloc[i_1][0], df_positive.iloc[i_2 + 1][0], weight))
                i_2 +=1
        
        else:
            i_2 +=1
               
    n += 1
    i_1 += 1
    i_2 = 0
```
